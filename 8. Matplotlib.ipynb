```python
import numpy as np

# make the data
#np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# Creazione dello scatter plot
plt.scatter(x, y, s=sizes, c=colors)


# Personalizzazione dell'asse x e y
plt.xlabel('Asse x')
plt.ylabel('Asse y')

# Aggiunta del titolo al grafico
plt.title('Scatter Plot')

# Visualizzazione del grafico
plt.show()
```

```python
from scipy.optimize import minimize_scalar
import numpy as np

# Definisci una funzione di cui trovare il minimo
def f(x):
    return (x+1)**2 + 3

# Trova il minimo
res = minimize_scalar(f)
print(res)
```

```python
x = np.arange(-6,5,0.1)
y = f(x)
plt.plot(x,y)
plt.title('Esempio calcolo minimo')
plt.plot(res.x,res.fun, 'ro')
```

Domanda 1 : linea di fit in uno scatter plot

ref : https://www.python-graph-gallery.com/scatterplot-with-regression-fit-in-matplotlib?utm_content=cmp-true

```python
# Generate data
x =  np.random.uniform(0, 10, size=100) # 100 valori tra 0 e 10
y = x +  np.random.normal(size=100) # Y = x, ma con un po' di rumore con media 0 


plt.figure(figsize=(16,9))
# Add scatterplot
plt.scatter(x, y, s=60)  

# Fit linear regression via least squares with numpy.polyfit
# It returns an slope (b) and intercept (a)
# deg=1 means linear fit (i.e. polynomial of degree 1)
b, a = np.polyfit(x, y, deg=1)

# Create sequence of 100 numbers from 0 to 100 
xseq = np.linspace(0, 10, num=100) # griglia per la visualizzazione

# Plot regression line
plt.plot(xseq, a + b * xseq, color="r", lw=2.5) # lw -> larghezza della linea
```
