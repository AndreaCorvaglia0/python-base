```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line plot

```python
x = np.arange(10)
y = x**2
```

```python
plt.plot(x,y)
plt.title('Esempio Line Plot')
plt.ylabel('y = x**2')
plt.xlabel('x')
plt.show()
```

```python
with plt.style.context('seaborn'):
    x = np.arange(1,4,0.1)

    # y = log(x)
    y = x
    plt.figure(figsize=(16,9))
    plt.plot(x,np.log(x), label='y = log(x)')

    # y = x
    y = x
    plt.plot(x,y, label= 'y = x')

    # y = x**2
    y = x
    plt.plot(x, x**2, label = r'$y=x^{2}$')

    # y = e**x
    y = x
    plt.plot(x,np.exp(x), label=r'$y=e^{x}$')

    plt.title('Esempio Line Plot multipli')
    plt.ylabel('y = x**2')
    plt.xlabel('x')
    plt.legend()
    plt.show()
```

## Distribuzioni

```python
np.random.seed(42)

# example data
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(10000)

num_bins = 1000
```

### Histogram

```python
from scipy.optimize import minimize_scalar
import numpy as np

# Definisci una funzione di cui trovare il minimo
def f(x):
    return (x+1)**2 + 3

# Trova il minimo
res = minimize_scalar(f)
print(res)
```

```python
x = np.arange(-6,5,0.1)
y = f(x)
plt.plot(x,y)
plt.title('Esempio calcolo minimo')
plt.plot(res.x,res.fun, 'ro')
```
